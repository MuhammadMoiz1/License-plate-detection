# Performance metrics

```python
def calculate_metrics(TP, FP, FN, TN):
    precision = TP / (TP + FP)
    recall = TP / (TP + FN)
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    f1_score = 2 * (precision * recall) / (precision + recall)
    return precision, recall, accuracy, f1_score


TP = 79
FP = 3
FN =  2
TN =  0

precision, recall, accuracy, f1_score = calculate_metrics(TP, FP, FN, TN)
print("Precision:", precision)
print("Recall:", recall)
print("Accuracy:", accuracy)
print("F1 Score:", f1_score)
```

```python
from ultralytics import YOLO
import cv2
import easyocr
import math
import os
reader = easyocr.Reader(['en'])

model = YOLO('best.pt')
model.export(format='onnx')

```

Function to get the text inside the bounding box ie: license plate

```python

def get_License_number(img_path):

  results=model(img_path)
  image = cv2.imread(img_path)
  if results and len(results[0].boxes.xyxy) > 0:
    # print(results[0].boxes)
    vals=results[0].boxes.xyxy[0].tolist()
    if len(vals)==4:

      xmin, ymin, xmax, ymax =  vals
      xmin = int(math.floor(xmin))
      ymin = int(math.floor(ymin))
      xmax= int(math.ceil(xmax))
      ymax = int(math.ceil(ymax))
      roi = image[ymin:ymax, xmin:xmax]
      result = reader.readtext(roi)
      cv2.rectangle(image, (xmin, ymin), (xmax, ymax), (0, 255, 0), 2)
      if result:
        text =  result[0][1]
        if text:
          cv2.putText(image, text, (xmin, ymin - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)



          if os.path.exists(img_path):
            os.remove(img_path)
            image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            cv2.imwrite(img_path, image_rgb)

  return
```

This portion of code will break your video into frames and then give those frames ptha to get_license_number function to exract text and then get_license _number function would replace the original frame with detected frame.

```python
import cv2
import os

# Create a folder to save frames
folder_name = "frames"
os.makedirs(folder_name, exist_ok=True)

video_path = vid
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Unable to open video.")
    exit()

# Initialize frame count
frame_count = 0

# Read until the end of the video
while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Check if the frame was read successfully
    if not ret:
        break

    # Save the frame as an image file

    frame_count= frame_count+1
    frame_path = os.path.join(folder_name, f"frame_{frame_count}.jpg")  # Save each frame in the folder
    cv2.imwrite(frame_path, frame)
    get_License_number(frame_path)

cap.release()
cv2.destroyAllWindows()

print(f"Total frames extracted: {frame_count}")
```

This   code would combine the frames into video again which would be saved with the name of output_video.mp4

---



```python
import cv2
import os

# Folder containing frames
folder_path = "frames"  # Replace with the path to your folder containing frames

# Output video file path
output_video_path = "output_video.mp4"  # Replace with the desired output video file path

# Get a list of frame image filenames sorted by their numeric order
frame_filenames = sorted([filename for filename in os.listdir(folder_path) if filename.endswith(".jpg") or filename.endswith(".png")], key=lambda x: int(x.split("_")[1].split(".")[0]))

# Get the frame width and height from the first image
first_image_path = os.path.join(folder_path, frame_filenames[0])
first_image = cv2.imread(first_image_path)
frame_width = first_image.shape[1]
frame_height = first_image.shape[0]

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # You can use other codecs like 'XVID', 'MJPG', 'X264', etc.
out = cv2.VideoWriter(output_video_path, fourcc, 25, (frame_width, frame_height))  # Adjust the frame rate as needed

# Iterate through each frame in the sorted list
for filename in frame_filenames:
    # Read the frame
    frame_path = os.path.join(folder_path, filename)
    frame = cv2.imread(frame_path)

    # Write the frame to the output video
    out.write(frame)

# Release the VideoWriter object
out.release()

print("Video conversion complete.")
```
